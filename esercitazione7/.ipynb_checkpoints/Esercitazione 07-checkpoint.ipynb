### 1.2
Da ora in poi, sino ad indicazione diversa, utilizziamo solo i primi 3 attributi come variabili esplicative. 

Addestrate un classificatore di tipo ``LogisticRegression`` (come visto a lezione) sul dataset di training e valutate i seguenti indici sul dataset di test: 

- Accuracy
- TP, FP, TN, FN (in questo dataset, intendiamo classe negativa la classe y=0)
- TPR, FPR

```python
# Solution

from sklearn.linear_model import LogisticRegression

def computePerformance(y_hat,y):
    # Inputs:
        # y_hat: vettore con m elementi (dove m è il numero di osservazioni) contente la classe predetta {-1,1} per ogni osservazione  
        # y: vettore con m elementi (dove m è il numero di osservazioni) contente la classe vera {-1,1} per ogni osservazione  
        # Outputs: True positive, Folse positive, True negative, False negative
    m=len(y)
    TN=0
    TP=0
    FN=0
    FP=0
    TPR=0
    FPR=0
    for i in range(m):
        if y[i]==y_hat[i] and y[i]==1:
            TP=TP+1
        elif y[i]==y_hat[i] and y[i]==0:
            TN=TN+1
        elif y[i]!=y_hat[i] and y[i]==1:
            FN=FN+1
        elif y[i]!=y_hat[i] and y[i]==0:
            FP=FP+1  
            
    print(FP,FN,TP,TN)
    Acc=((TP+TN)/(m))
    TPR=TP/(TP+FN)
    FPR=FP/(FP+TN)
    
    return Acc, TP, FP, TN, FN, TPR, FPR


n = 3 # Numero di features considerate
clf = LogisticRegression()
clf.fit(xTrain[:,0:n], yTrain)
y_hat = clf.predict(xTest[:,0:n])

Acc, TP, FP, TN, FN, TPR, FPR = computePerformance(y_hat,yTest)
print(f"Accuracy: {Acc}")
print(f"TP: {TP}, FP: {FP}, TN: {TN}, FN:{FN}")
print(f"TPR: {TPR}, FPR: {FPR}")
```

### 1.4
Si ripeta l'esercizio 2.4 dell'esercitazione precedente. Nello specifico, scrivete una funzione python ``computeTPRandFPR`` che calcoli i valori del TPR e FPR al variare della soglia t del classificatore (l'osservazione è attribuita alla classe 1 se la probabilità di appartenere alla classe 1 è maggiore di t). La funzione riceve in ingresso un vettore y con la classe vera della osservazioni, un vettore p con lo score del classificatore (probabilità di appartenere alla classe 1), t: soglia del classificatore. 

Potete ovviamente riutilizzare la funzione scritta nell'esercitazione precedente, assicurandovi pero' che le istanze con classe negativa siano tali che y = 0 e non y = -1.

Verificate le risposte alle domande del punto precedente.

Hint: Come visto a lezione, lo score della classe 0 e della classe 1 si ottiene col comando ``clf.predict_proba``. Nota che a noi interessa solo lo score della classe 1!!

```python
# Solution 
from sklearn.linear_model import LogisticRegression

def computeTPRandFPR(y,p,t):
    # Inputs:
        # y: vettore con m elementi (dove m è il numero di osservazioni) contente la classe vera {-1,1} per ogni osservazione  
        # p: vettore con m elementi (dove m è il numero di osservazioni) contente la probabilità di appartenere alla classe 1
        # t: soglia (valore tra 0 e 1)    
    m=len(y)
    TN=0
    TP=0
    FN=0
    FP=0
    for i in range(m):
        if y[i]==1 and p[i]>=t:
            TP=TP+1
        elif y[i]==-1 and p[i]<t:
            TN=TN+1
        if y[i]==1 and p[i]<t:
            FN=FN+1
        elif y[i]==-1 and p[i]>=t:
            FP=FP+1
        
    if(FN+TP==0):
        print(f"division by zeros !!")
        TPR=0
    else:
        TPR=TP/(TP+FN)
    if(FP+TN==0):
        print(f"divisioni by zero !!")
        FPR=0
    else:
        FPR=FP/(FP+TN)
    return TPR, FPR


for t in [0.1, 0.5, 0.9]:
    y = np.array([1,1,1,1,1,1,-1,-1])  
    p = np.array([0.9,0.85,0.8,0.7,0.5,0.4,0.6,0.1]) 
    (TPR, FPR) = computeTPRandFPR(y,p,t)

    print(f"True positive Rate: {TPR}")
    print(f"False positive Rate: {FPR}")
   
```
