Se proviamo ad eseguire i comandi della cella sopra, il server HTTP ci restituisce un errore di tipo ``HTTP Error 403: Forbidden`` (in parole povere, il server HTTP non ha accettato la richiesta di fare il parsing del file HTML). Senza entrare nel dettaglio, facciamo in modo che la nostra richiesta al server venga inviata come verrebbe inviata tramite un browser. Per fare questo basta eseguire i comandi della cella sotto. 

```python
import pandas as pd
import requests

url = 'https://www.fifa.com/worldcup/statistics/players/shots'

header = {
  "User-Agent": "Chrome/50.1.3672",
  "X-Requested-With": "XMLHttpRequest"
}

r = requests.get(url, headers=header)
print(r.text)
listaDf = pd.read_html(r.text)
print(listaDf)
df = listaDf[0]
```

### 1.1
Si generi un nuovo array (che chiameremo arrayShotsPerGame) che calcoli, per ogni giocatore, il numero di tiri a partita (dove la partita si intende $90$ minuti di gioco). 

Esempio 1: Il giocatore *Fuoriclasse* ha giocato durante il mondiale 135 minuti (una partita e mezzo) e ha effettuato 19 tiri, allora il numero di tiri a partita è uguale a
$19*\frac{90}{135} = 12.66$. 

Esempio 2: Il giocatore *Campione* ha giocato durante il mondiale solo 5 minuti e ha effettuato $4$ tiri, allora il numero di tiri a partita è uguale a
$4*\frac{90}{5} = 72$!!! (Durante il mondiale ha tirato solo $4$ volte, ma ha giocato molto poco!!!!). 

Hint: il numero di minuti giocati da ogni giocatore si trova nell colonna con label ``Minutes Played  MinP``.


```python
playedTime=df['Minutes Played  MinP'].values
arrayShotsPerGame=arrayShots*90/playedTime
print(arrayShotsPerGame)
```

### 1.4
Calcolare lo z-score delle osservazioni e identificate gli outliers. Sapete spiegare la ragione di questi outliers?

```python
#Calcolare lo z-score delle osservazioni e identificate gli outliers. Sapete spiegare la ragione di questi outliers?
#z-scores
zs=(arrayShotsPerGame - np.mean(arrayShotsPerGame)) / np.std(arrayShotsPerGame)
#print("Array di z-Score")
#print(zs)

mask= (zs<3) & (zs>-3)
#print(mask)
#print(np.sum(mask))
#arraysenzaoutliers
arrays=arrayShotsPerGame[mask]
#print(arrays)

#la ragione degli outsider secondo è a causa dei calci di rigore in quanto vi sono giocatori che 
#probabilmente hanno svolto solo quella parte di gara e quindi hanno avuto un numero di tiri molto alto
#pur avendo giocato molto poco
```
