Si consideri la variabile numerica ``età`` e la variabile categorica ``Lavoratore`` che assume valori "YES" (se l'utente dichiara di avere un lavoro), "NO" (se l'utente dichiara di non avere un lavoro). La cella sotto esclude dal dataset tutti gli utenti che sono studenti o che non dichiarano il proprio lavoro e raggruppa gli utenti in base alla variabile ``Lavoratore`` definita sopra. L'array finale è ``jobArray``. Opzionale: Provate a capire i comandi eseguiti nella cella sotto.  

```python
dfNew = df.dropna(subset = ['job'])

m = len(dfNew['job'].values)
jobArray = np.full(m, "", dtype = object)

for i in range(m):
    if dfNew['job'].values[i] == 'unemployed':
        jobArray[i] = 'NO' 
    else:
        jobArray[i] = 'YES' 
```

### 2.1 
Disegnare un grouped boxplot che confronti la distribuzione dell'età per le varie sottoclassi identificate dalla variabile ``Lavoratore``.  

```python
# Solution
# Rappresentazione come boxplot
support=np.unique(jobArray)
eta=dfNew['age'].values
data = []
for val in support:
    num_g = eta[jobArray == val] # tempi per le osservazioni corrispondenti
    trace = go.Box(x=num_g, name=val)
    data.append(trace)

layout = go.Layout(
    xaxis=dict(title="Età"),
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```

```python
# Solution



# Solution
data = []
for val in support:
    num_g = eta[jobArray == val] # tempi per le osservazioni corrispondenti
    trace = go.Histogram(x=num_g, name=val)
    data.append(trace)

layout = go.Layout(
    barmode='group',
    xaxis=dict(title="Età"),
    yaxis=dict(title="Numero persone"),
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
#perchè il numero di persone che non hanno un lavoro per una determinata età sono molto meno rispetto a quelle che hanno lavoro e quindi l'istogramma del NO risulta poco visibile
```

### 2.3 (opzionale)

L'istogramma generato al punto precedente che rappresenta la distribuzione delle età degli utenti disoccupati non è ben visibile nella stessa figura dell'istogramma che rappresenta la distribuzione delle età degli utenti lavoratori poichè il numero di utenti disoccupati è molto piu' basso rispetto ai lavoratori. Di conseguenza, anche le barre dell'istogramma sono molto piu' basse rispetto alle barre dell'istogramma caratterizzante gli utenti lavoratori. In questo caso, sarebbe meglio "normalizzare" gli istogrammi, dove l'altezza delle barre non è la frequenza assoluta (numero di utenti in un intervallo di età), ma la frequenta relativa (frazione di utenti in un intervallo di età). La cella sotto mostra come generare un istogramma normalizzato (basta impostare l'attributo ``histnorm`` uguale a ``probability'`` quando si creare l'oggetto della classe ``go.Histogram``).  

```python
#Hint
```

Generate adesso gli istogrammi normalizzati dell'età dei disoccupati e dei lavoratori.

```python
# Solution


# Solution
data = []
for val in support:
    print(val)
    num_g = eta[jobArray == val] # tempi per le osservazioni corrispondenti
    trace = go.Histogram(x=num_g, name=val,histnorm = 'probability')
    data.append(trace)

layout = go.Layout(
    barmode='group',
    xaxis=dict(title="Età"),
    yaxis=dict(title="Numero persone"),
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```
